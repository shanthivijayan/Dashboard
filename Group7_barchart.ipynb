```python
#import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading the data set
filepath='/content/supermarket_sales.csv'
df_supermarket=pd.read_csv(filepath)
df_supermarket.head()
```

```python
df_supermarket.info()
df_supermarket.describe()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df_supermarket, x='Product line')
plt.title('Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
dfplot=df_supermarket['Product line'].value_counts()
dfplot.plot(kind='bar')
```

World population data

```python
df_population=pd.read_csv('/content/world_population.csv')
df_population.head()
```

```python
#using sns
plt.figure(figsize=(10, 6))
sns.countplot(data=df_population, x='Continent')
plt.title('Population by Continent')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```

```python
#using matplotlib
df_contient=df_population.loc[:'Continent','2022 Population']
df_contient1=df_population['Continent'].value_counts()
df_contient1.plot(kind='barh')
```
